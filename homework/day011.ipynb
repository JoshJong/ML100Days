```python
#根據題目給的 DataFrame 完成下列操作：
#- 計算每個不同種類 animal 的 age 的平均數
#- 將資料依照 Age 欄位由小到大排序，再依照 visits 欄位由大到小排序
#- 將 priority 欄位中的 yes 和 no 字串，換成是布林值 的 True 和 False
import numpy as np
import pandas as pd
data = {
    'animal': ['cat', 'cat', 'snake', 'dog', 'dog', 'cat', 'snake', 'cat', 'dog', 'dog'],
    'age': [2.5, 3, 0.5, np.nan, 5, 2, 4.5, np.nan, 7, 3],
    'visits': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
    'priority': ['yes', 'yes', 'no', 'yes', 'no', 'no', 'no', 'yes', 'no', 'no']
}

labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
df = pd.DataFrame(data, index=labels)
df
```

```python
df.sort_values(by=['animal'])
```

```python
for item in df['animal'].unique():
    print(item, df[df['animal']==item]['age'].mean())
    #print(' ')
```

```python
df.dropna().sort_values(by=['age','visits'], ascending=[True, False])
```

```python
df['priority'] = (df['priority']=='yes')
df
```

```python
#一個包含兩個欄位的 DataFrame，將每個數字減去
#* 1) 該欄位的平均數
#* 2) 該筆資料平均數
df = pd.DataFrame(np.random.random(size=(5, 3)))
print(df)
```

```python
print(df.mean(axis=0))
df1 = df - df.mean(axis=0)
df1
```

```python
print(df.mean(axis=1))
df.sub(df.mean(axis=1), axis=0)
```
