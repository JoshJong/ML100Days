```python
#請比較對一個 1000 x 100 x 100 的陣列，使用不同方法對每一個元素 +1 的時間比較。
import random

def function1():
    np.random.seed(0)
    a = np.random.randint(0,100,1000*100*100)   
    #print(a)
    a.resize([1000,100,100])
    #print(a)
    return a+1
```

```python
%timeit [print(function1().flatten())]
```

```python
def function2():
    np.random.seed(0)
    a = np.random.randint(0,100,1000*100*100)  
    a.resize([1000,100,100]) 
    for i in range(len(a)):
        for j in range(len(a[i])):
            for k in range(len(a[i][j])):
                a[i][j][k]+=1
    return a
```

```python
%timeit [print(function2().flatten())]
```

```python
#如何利用 list(...) 實現 a.tolist() 的效果？試著用程式實作。
import numpy as np

#np.random.seed(0)
a = np.random.randint(0,100,100)   
b = a
c = a.reshape([2,50])
d = a.reshape(2,25,2)

print(b)
print(c) 
print(d)
```

```python
print(b.tolist())
print(c.tolist()) 
print(d.tolist())

print(list(b)) #elements are ndarrays
print(list(c)) #elements are ndarrays
print(list(d)) #elements are ndarrays
```

```python
def myfunction(ndarray):
    if type(ndarray) != np.ndarray:
        return ndarray

    #for i in np.nditer(ndarray):
    nd_list=list(ndarray)
    for i in range(len(nd_list)):
        nd_list[i] = myfunction(nd_list[i])
    return nd_list

print(myfunction(b))
print(myfunction(c))
print(myfunction(d))
```

```python
#如何從一個陣列中，找出出現頻率最高的數值與位置？
import numpy as np

np.random.seed(0)
a = np.random.randint(0,10,50)
print(a)

counts = np.bincount(a)
print(counts)  #how many elements from 0,1,2,3,4,5...
print("most frequent number is : " + str(np.argmax(counts)))
```
