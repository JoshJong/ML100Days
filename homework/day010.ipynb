```python
# 請根據提供的資料，選擇出下列的要求：
# - 1. filtered by first column > 20?
# - 2. filtered by first column + second column > 50
# - 3. filtered by first column < 30 or second column > 30
# - 4. filtered by total sum of row > 100
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randint(10, 40, 60).reshape(-1, 4))
print(df)
print()

print(df[df.iloc[0:,0:1]>20]) #1
print(df[df.iloc[0:,0:2]>50]) #2
print(pd.merge(df[df.iloc[0:,0:1]<30].iloc[0:,0:1] , df[df.iloc[0:,1:2]>30].iloc[0:,1:2], on=df.index, how='outer')) #3


df['sum'] = df.sum(axis=1)
print(df)
print(df[df['sum'].iloc[0:]>100].iloc[0:,:-1]) #4
```
